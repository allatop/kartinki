# **Задание 2. Формирование иерархической структуры проекта в контексте управления проектами информационных систем**






## **Введение**

 **Структурированное планирование является фундаментальным навыком проектного менеджера**. Создание иерархической структуры работ (WBS - Work Breakdown Structure) - это один из основных инструментов управления проектами, который позволяет декомпозировать крупные задачи на управляемые компоненты.

## **Формулировка задания**

На основе ваших ответов на предыдущие вопросы о цели, задачах и практической значимости проекта ([Задание №1](https://colab.research.google.com/drive/12wuBTgZlyS8ZYLXjw2SQ2F-wHGbBU0nC?usp=sharing)), сформулируйте его детальную структуру, следуя приведенному шаблону:

1. Структура должна содержать **минимум** 3 раздела первого уровня (главы)
2. Каждый раздел первого уровня должен содержать **минимум** 3 подраздела второго уровня (параграфы)
3. Каждый подраздел второго уровня должен содержать **минимум** 3 подраздела третьего уровня (подпараграфы)

Для каждого элемента структуры предоставьте краткое описание содержания (1-2 предложения), которое поясняет, что будет рассмотрено в данном разделе и как это соотносится с общей целью и задачами работы.

#### **Пример №1. Типовая структура исследовательского проекта**


```
1. ТЕОРЕТИЧЕСКИЕ ОСНОВЫ И ОБЗОР ПРЕДМЕТНОЙ ОБЛАСТИ
   1.1. Современное состояние исследуемой проблематики
      1.1.1. Текущий уровень и состояние данной области
      1.1.2. Современные подходы и методы решения рассматриваемой задачи
      1.1.3. Основные проблемы существующих решений

   1.2. Анализ теоретической и методологической базы
      1.2.1. Фундаментальные принципы и подходы в области исследования
      1.2.2. Современные методы и технологии в данной сфере
      1.2.3. Нормативная база и стандарты (при наличии)

   1.3. Обзор существующих решений
      1.3.1. Отечественный опыт решения аналогичных задач
      1.3.2. Зарубежные исследования и разработки


2. АНАЛИЗ ПРОБЛЕМАТИКИ И ОБОСНОВАНИЕ ПРЕДЛАГАЕМОГО РЕШЕНИЯ
   2.1. Детальный анализ объекта исследования
      2.1.1. Анализ достоинств и недостатков существующих подходов
      2.1.2. Выявление основных проблем
      2.1.3. Определение направлений совершенствования

   2.2. Формирование концепции авторского решения
      2.2.1. Обоснование выбранного подхода и методологии
      2.2.2. Формулировка требований к разрабатываемому решению
      2.2.3. Ожидаемые результаты и критерии их оценки

   2.3. Построение модели предлагаемого решения
      2.3.1. Концептуальная модель или схема решения
      2.3.2. Основные компоненты и их взаимосвязи
      2.3.3. Условия и границы применимости решения

3. ПРАКТИЧЕСКАЯ РЕАЛИЗАЦИЯ И ОЦЕНКА РЕЗУЛЬТАТОВ
   3.1. Реализация предложенного решения
      3.1.1. Описание этапов разработки или внедрения
      3.1.2. Инструментарий и ресурсы для реализации
      3.1.3. Основные особенности практической реализации

   3.2. Апробация и верификация результатов
      3.2.1. Методика проведения экспериментов или испытаний
      3.2.2. Анализ полученных результатов
      3.2.3. Сопоставление с исходными требованиями и ожиданиями

   3.3. Оценка эффективности и перспективы развития
      3.3.1. Достигнутые показатели и их интерпретация
      3.3.2. Рекомендации по практическому применению
      3.3.3. Направления дальнейших исследований и улучшений

```


#### **Пример №2. Специализированная структура IT-проекта**

```
1. ОБЗОР ПРЕДМЕТНОЙ ОБЛАСТИ И СУЩЕСТВУЮЩИХ РЕШЕНИЙ
   1.1. Исследование текущего состояния предметной области
      1.1.1. Современные тенденции и проблемы в исследуемой сфере
      1.1.2. Существующие подходы к решению выявленных проблем

   1.2. Обзор существующих систем и приложений
      1.2.1. Отечественные решения и их функциональные возможности
      1.2.2. Зарубежные аналоги и опыт их применения

   1.3. Обзор технологий и инструментов разработки
      1.3.1. Современные языки программирования и фреймворки
      1.3.2. Платформы и среды разработки
      1.3.3. Подходы к созданию аналогичных решений (типовые стеки разработки)

2. АНАЛИЗ И ФОРМИРОВАНИЕ ТРЕБОВАНИЙ К СИСТЕМЕ
   2.1. Анализ недостатков существующих решений
      2.1.1. Выявление функциональных ограничений аналогов
      2.1.2. Проблемы производительности, масштабируемости, ограниченной области применения

   2.2. Выбор технологического стека на основе анализа
      2.2.1. Обоснование выбора языков программирования
      2.2.2. Определение архитектуры и платформы развёртывания
      2.2.3. Выбор инструментов разработки и тестирования

   2.3. Формирование требований к разрабатываемой системе
      2.3.1. Функциональные требования и пользовательские сценарии (диаграмма вариантов использования, диаграмма деятельности)
      2.3.2. Нефункциональные требования
      2.3.3. Критерии оценки эффективности решения (индивидуально)

3. ПРОЕКТИРОВАНИЕ И РЕАЛИЗАЦИЯ СИСТЕМЫ
   3.1. Проектирование архитектуры приложения
      3.1.1. Проектирование структуры базы данных (ER-диаграмма)
      3.1.2. Проектирование модулей и их взаимосвязей (Диаграмма классов)
      3.1.3. UML-диаграммы и схемы взаимодействия компонентов (Диаграмма компонентов, диаграмма автоматов, диаграмма деятельности и т.д.)

   3.2. Программная реализация системы
      3.2.1. Разработка серверной части и бизнес-логики
      3.2.2. Создание пользовательского интерфейса
      3.2.3. Интеграция с внешними системами и API

   3.3. Внедрение и тестирование
      3.3.1. Развёртывание системы в рабочем окружении
      3.3.2. Функциональное и нагрузочное тестирование (индивидуально)

```
